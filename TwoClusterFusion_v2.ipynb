```python
import numpy as np
import matplotlib.pyplot as plt

font = {'family' : 'Arial',
        'size'   : 16}

plt.rc('font', **font)

file1 = "./Two_cluster_fusion_S10N03kT/Mean_CVtrajectory.txt"

d = np.loadtxt(file1)

plt.plot(d[:,0]/1e6, d[:,1], lw=2)
plt.fill_between(d[:,0]/1e6, d[:,1]-d[:,2], d[:,1]+d[:,2], alpha=0.3)
plt.ax

plt.ylabel('R$_{dist}$')
plt.xlabel('TimeSteps ($10^6$)')

plt.show()
```
