## Timecourse of Collective Variable (order parameter) 
Average of 5 stochastic traj (Run*.colvars.traj) files which is saved as "Mean_CVtrajectory.txt"

```python
import numpy as np
import matplotlib.pyplot as plt

font = {'family' : 'Arial',
        'size'   : 16}

plt.rc('font', **font)

file1 = "./N200_Clustering/MultiRun_data/Mean_CVtrajectory.txt"

d = np.loadtxt(file1)

plt.plot(d[:,0]/1e6, d[:,1], lw=2)
plt.fill_between(d[:,0]/1e6, d[:,1]-d[:,2], d[:,1]+d[:,2], alpha=0.3)

plt.ylim(80,160)
plt.ylabel('R$_g ^{system}$ ($\AA$)')
plt.xlabel('TimeSteps (10$^6$)')
plt.show()
```

```python
file2 = "./N200_Clustering/MultiRun_data/Mean_PMF.txt"

d = np.loadtxt(file2)

plt.plot(d[:,0], d[:,1], lw=2)
plt.fill_between(d[:,0], d[:,1]-d[:,2], d[:,1]+d[:,2], alpha=0.3)

plt.xlabel('R$_g ^{system}$ ($\AA$)')
plt.ylabel('PMF (kcal/mol)', fontweight='bold', fontsize=14)
plt.show()
```
